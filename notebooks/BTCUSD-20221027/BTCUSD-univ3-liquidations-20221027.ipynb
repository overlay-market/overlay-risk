```python
import pandas as pd
```

```python
FILEPATH = "../../csv/metrics/BTCUSD-600-20210401-20210630"
```

```python
FILENAME_MMS = f"{FILEPATH}-mms.csv"
FILENAME_BETAS = f"{FILEPATH}-betas.csv"
```

```python
df_mms = pd.read_csv(FILENAME_MMS)
df_mms
```

```python
# Change Unnamed: 0 column name and convert secs to hours
df_mms['hours'] = df_mms['Unnamed: 0']
df_mms.drop('Unnamed: 0', axis=1, inplace=True)
df_mms['hours'] = df_mms['hours'].apply(lambda x: int(x[2:])/3600)
df_mms
```

```python
# MM value at 4 hours for 95% confidence
df_mms[df_mms['hours'] == 4]['alpha=0.05']
```

```python
# MM value at 2 days for 95% confidence
df_mms[df_mms['hours'] == 48]['alpha=0.05']
```

27% is impractical. 4.5% is practical but might not be safe to expect liquidations within 4 hours (on mainnet, given gas spikes that last days).
