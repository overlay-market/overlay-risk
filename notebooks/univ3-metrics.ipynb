```python
# arrange a separate df for alpha param to create pretty plot
df_wethusdc_pc1_alphas = df_wethusdc_pc1[df_wethusdc_pc1['_field'] == 'alpha'].filter(['_time', '_value'])
df_wethusdc_pc1_alphas.columns = [ 'time', 'a' ]
df_wethusdc_pc1_alphas.set_index('time')
```

```python
# save pretty plot for alpha
df_alphas_plt = df_wethusdc_pc1_alphas.plot(x='time', y='a', color='tab:blue')
df_alphas_plt.get_figure().savefig('img/pstable-wethusdc-alpha.png')
df_alphas_plt.get_figure().savefig('img/pstable-wethusdc-alpha.eps')
```

```python
df_wethusdc_pc1_alphas['a'].mean()
```

```python
df_wethusdc_pc1_alphas['a'].std()
```

```python
df_wethusdc_pc1_alphas['a'].std()/df_wethusdc_pc1_alphas['a'].mean()
```

```python
# arrange a separate df for beta param to create pretty plot
df_wethusdc_pc1_betas = df_wethusdc_pc1[df_wethusdc_pc1['_field'] == 'beta'].filter(['_time', '_value'])
df_wethusdc_pc1_betas.columns = [ 'time', 'b' ]
df_wethusdc_pc1_betas.set_index('time')
```

```python
# save pretty plot for beta
df_betas_plt = df_wethusdc_pc1_betas.plot(x='time', y='b', color='tab:purple')
df_betas_plt.get_figure().savefig('img/pstable-wethusdc-beta.png')
df_betas_plt.get_figure().savefig('img/pstable-wethusdc-beta.eps')
```

```python
# arrange a separate df for mu param to create pretty plot
df_wethusdc_pc1_mus = df_wethusdc_pc1[df_wethusdc_pc1['_field'] == 'mu'].filter(['_time', '_value'])
df_wethusdc_pc1_mus.columns = [ 'time', 'mu * tau' ]
df_wethusdc_pc1_mus.set_index('time')
```

```python
# save pretty plot for mu
df_mus_plt = df_wethusdc_pc1_mus.plot(x='time', y='mu * tau', color='tab:gray')
df_mus_plt.get_figure().savefig('img/pstable-wethusdc-mu.png')
df_mus_plt.get_figure().savefig('img/pstable-wethusdc-mu.eps')
```

```python
# arrange a separate df for sigma param to create pretty plot
df_wethusdc_pc1_sigmas = df_wethusdc_pc1[df_wethusdc_pc1['_field'] == 'sigma'].filter(['_time', '_value'])
df_wethusdc_pc1_sigmas.columns = [ 'time', 'sigma * (tau/a)**(1/a)' ]
df_wethusdc_pc1_sigmas.set_index('time')
```

```python
df_wethusdc_pc1_sigmas['sigma * (tau/a)**(1/a)']
```

```python
df_wethusdc_pc1_sigmas['sigma * (tau/a)**(1/a)'].mean()
```

```python
df_wethusdc_pc1_sigmas['sigma * (tau/a)**(1/a)'].std()
```

```python
df_wethusdc_pc1_sigmas['sigma * (tau/a)**(1/a)'].std() / df_wethusdc_pc1_sigmas['sigma * (tau/a)**(1/a)'].mean()
```

```python
# save pretty plot for sigma
df_sigmas_plt = df_wethusdc_pc1_sigmas.plot(x='time', y='sigma * (tau/a)**(1/a)', color='tab:green')
df_sigmas_plt.get_figure().savefig('img/pstable-wethusdc-sigma.png')
df_sigmas_plt.get_figure().savefig('img/pstable-wethusdc-sigma.eps')
```
